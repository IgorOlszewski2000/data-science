Wyznacznik macierzy

```python
A = np.array([[2,4], [-1,3]])
round(np.linalg.det(A))
```

```python
A = np.array([[2,4], [-1,3]])

np.trace(A)
```

Macierz odwrotna

```python
B = np.linalg.inv(A)
B
```

```python
np.set_printoptions(precision=4, suppress=True)
np.dot(A,B)
np.dot(B,A)
```
