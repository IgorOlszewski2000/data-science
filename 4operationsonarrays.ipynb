```python
import numpy as np
```

```python
A = np.array([3, 1, 4, 2])
B = np.array([3, -1, 3, 2])
print(A)
print(B)
```

```python
A + B
```

```python
A / B
```

```python
A + 3
```

```python
2 * A
```

```python
np.add(A, B)
```

```python
np.subtract(A,B)
```

```python
np.divide(A,B)
```

```python
np.multiply(A,B)
```

```python
X = np.array([[1,3] , [-2,0]])
Y = np.array([[6,0] , [-1,2]])
print(X)
print(Y)
```

```python
np.dot(X,Y) # mnozenie maciezowe
```

```python
X.dot(Y)
```

```python
X @ Y # tez mnozenie maciezowe
```
